```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
#label= ["DT", "RF",'XG','AD','CT']
    
data_2017 = [1.0, 1.0, 0.97, 1.0, 1.0]
data_2018 = [1.0, 1.0, 1.0, 0.98, 1.0]
labels = ['Decision Tree','Random Forest','XGBoost','AdaBoost','CatBoost']

# Set the width of each bar
bar_width = 0.4

# Calculate the x positions of the bars
x_2017 = np.arange(len(labels))
x_2018 = x_2017 + bar_width

# Create the figure and axis objects
fig, ax = plt.subplots()

# Plot the bars
ax.bar(x_2017, data_2017, width=bar_width, label='CIC-IDS2017')
ax.bar(x_2018, data_2018, width=bar_width, label='NSL-KDD')

# Add labels, title, and legend
ax.set_ylabel('Accouracy')
ax.set_xticks(x_2017 + bar_width / 2)
ax.set_xticklabels(labels, rotation=45)
ax.legend()
#
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
# Add numbers as text labels above the bars
for i, v in enumerate(data_2017):
    ax.text(x_2017[i] - 0.15, v + 0.05, '{:.2f}'.format(round(v, 2)), color='black', fontsize=7, ha='center')
for i, v in enumerate(data_2018):
    ax.text(x_2018[i] - 0.15, v + 0.05, '{:.2f}'.format(round(v, 2)), color='black',  fontsize=7, ha='center')

# Show the plot
plt.tight_layout()
plt.show()
```
